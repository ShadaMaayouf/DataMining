### Streudiagramme <a name=d></a>

```python
# DataFrame erstellen
df_mitarbeiter = pd.DataFrame({
    'Alter': [26, 34, 32, 21, 27, 39, 41, 30, 45, 37, 46, 35, 39, 52, 51, 54],
    'Gehalt': [2.1, 2.3, 2.4, 2.5, 2.5, 2.8, 2.8, 3.0, 3.2, 3.4, 3.5, 3.8, 4.8, 5.0, 5.7, 5.9],
    'Abteilung': ['Vertrieb', 'Marketing', 'IT', 'Vertrieb', 'IT', 'Marketing', 'Vertrieb', 'IT', 'Marketing', 'Vertrieb', 'IT', 'Marketing', 'Vertrieb', 'IT', 'Marketing', 'Vertrieb']
})
```

```python
import matplotlib.pyplot as plt

# Streudiagramm erstellen
plt.scatter(df_mitarbeiter['Alter'], df_mitarbeiter['Gehalt'])
plt.title('Streudiagramm von Alter und Gehalt')
plt.xlabel('Alter')
plt.ylabel('Gehalt (in Tausend Euro)')
plt.show()
```

## 2.3.2. Visualisierung von Mehrdimensionalen Daten

### Streudiagramme <a name=2a></a>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten in einem DataFrame speichern

# Farben für jede Abteilung
farben = {'Vertrieb': 'red', 'Marketing': 'blue', 'IT': 'green'}

# Streudiagramm erstellen
for abt in df_mitarbeiter['Abteilung'].unique():
    plt.scatter(
        df_mitarbeiter[df_mitarbeiter['Abteilung'] == abt]['Alter'], 
        df_mitarbeiter[df_mitarbeiter['Abteilung'] == abt]['Gehalt'], 
        c = farben[abt], label = abt, s = 100
    )

plt.title('Streudiagramm von Alter, Gehalt und Abteilung')
plt.xlabel('Alter')
plt.ylabel('Gehalt (in Tausend Euro)')
plt.legend()
plt.show()
```

### Streudiagramm-Matrix <a name=2b></a>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten in einem DataFrame speichern
df = pd.DataFrame({
    'Alter': [26, 34, 32, 21, 27, 39, 41, 30, 45, 37, 46, 35, 39, 52, 51, 54],
    'Gehalt': [2.1, 2.3, 2.4, 2.5, 2.5, 2.8, 2.8, 3.0, 3.2, 3.4, 3.5, 3.8, 4.8, 5.0, 5.7, 5.9],
    'Abteilung': ['Vertrieb', 'Marketing', 'IT', 'Vertrieb', 'IT', 'Marketing', 'Vertrieb', 'IT', 'Marketing', 'Vertrieb', 'IT', 'Marketing', 'Vertrieb', 'IT', 'Marketing', 'Vertrieb']
})

# Farben für jede Abteilung
farben = {'Vertrieb': 'red', 'Marketing': 'blue', 'IT': 'green'}

# Streudiagramm-Matrix erstellen
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Alter vs Gehalt
for abt in df['Abteilung'].unique():
    axs[0, 0].scatter(df[df['Abteilung'] == abt]['Alter'], df[df['Abteilung'] == abt]['Gehalt'], c = farben[abt], label = abt)
axs[0, 0].set_xlabel('Alter')
axs[0, 0].set_ylabel('Gehalt')

# Alter vs Abteilung
for abt in df['Abteilung'].unique():
    axs[0, 1].scatter(df[df['Abteilung'] == abt]['Alter'], df[df['Abteilung'] == abt]['Abteilung'], c = farben[abt], label = abt)
axs[0, 1].set_xlabel('Alter')
axs[0, 1].set_ylabel('Abteilung')

# Gehalt vs Alter
for abt in df['Abteilung'].unique():
    axs[1, 0].scatter(df[df['Abteilung'] == abt]['Gehalt'], df[df['Abteilung'] == abt]['Alter'], c = farben[abt], label = abt)
axs[1, 0].set_xlabel('Gehalt')
axs[1, 0].set_ylabel('Alter')

# Gehalt vs Abteilung
for abt in df['Abteilung'].unique():
    axs[1, 1].scatter(df[df['Abteilung'] == abt]['Gehalt'], df[df['Abteilung'] == abt]['Abteilung'], c = farben[abt], label = abt)
axs[1, 1].set_xlabel('Gehalt')
axs[1, 1].set_ylabel('Abteilung')

plt.legend()
plt.show()
```
